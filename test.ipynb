```python
import yaml
from sqlalchemy import create_engine
from sqlalchemy import inspect
import pandas as pd


with open('../Multinational Retail Data Centralisation/db_creds.yaml' , 'r') as db_creds_file:
    db_creds = yaml.safe_load(db_creds_file)
print(db_creds)
engine = create_engine(f"{db_creds['DATABASE_TYPE']}+{db_creds['DBAPI']}://{db_creds['RDS_USER']}:{db_creds['RDS_PASSWORD']}@{db_creds['RDS_HOST']}:{db_creds['RDS_PORT']}/{db_creds['RDS_DATABASE']}")
print(engine)

df = pd.read_sql_table('legacy_users', engine)

users = df

users.country_code.unique()
substring = "NULL"
users[users.apply(lambda row: row.astype(str).str.contains(substring, case=False).any(), axis=1)]

import numpy as np
users = users.replace ('NULL', np.nan)
users[users.apply(lambda row: row.astype(str).str.contains(substring, case=False).any(), axis=1)]


countries = ['United Kingdom', 'United States', 'Germany']
users['country'] = users['country'].apply(lambda x: x if x in countries else np.nan)





```

```python
users["date_of_birth"] = pd.to_datetime(users["date_of_birth"], infer_datetime_format=True, errors='coerce') 
users['date_of_birth'].head(10)
users['join_date'] = pd.to_datetime(users['join_date'], infer_datetime_format=True, errors='coerce')
users['join_date'].head(10)


```
